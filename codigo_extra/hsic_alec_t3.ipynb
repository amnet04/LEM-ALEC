```python
import csv
import numpy
import pandas
from hsic import HSIC
```

```python
origen_varientes = "/src/media/supermatrix.csv"
destino = "/src/media/hsic_samp5000.csv"
geo=[]
reader = pandas.read_csv(origen_varientes, chunksize=1, index_col=0, sep ="\t", quotechar="'")
count = 0

for chunck  in reader:
    if count < 2:
        geo.append(chunck.drop('mapa', axis=1).transpose())
    if count == 1:
        geo = numpy.column_stack((geo[0],geo[1]))
    if count >= 2:
        data = chunck.drop('mapa', axis=1).as_matrix().transpose()
        variante = chunck.index[0]
        mapa = (chunck['mapa'][0])
        hsic = HSIC.HSIC_pval(geo,data,kernelX="Gaussian", kernelY="Delta", N_samp=5000)
        if count == 2:
            with open(destino, 'w') as csvfile:
                fieldnames = ['variante','mapa','HSIC','PValue']
                writer = csv.DictWriter(csvfile, fieldnames=fieldnames, dialect='excel-tab')
                writer.writeheader()
                writer.writerow({
                    'variante':variante,
                    'mapa':mapa,
                    'HSIC':hsic[0],
                    'PValue':hsic[1],
                })
        else:
            with open(destino, 'a') as csvfile:
                writer = csv.DictWriter(csvfile, fieldnames=fieldnames, dialect='excel-tab')
                writer.writerow({
                    'variante':variante,
                    'mapa':mapa,
                    'HSIC':numpy.float64(hsic[0]),
                    'PValue':numpy.float64(hsic[1]),
                })   
        print ('\r {}:{}'.format(count-1,variante), end="\t \t \t \t \t \t")
    count+=1
```
